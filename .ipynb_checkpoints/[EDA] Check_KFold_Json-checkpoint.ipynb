```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from PIL import Image
```

```python
data_folder = '../data'
with open(os.path.join(data_folder, 'cv_train_1.json')) as file:
   train_json = json.load(file)
with open(os.path.join(data_folder, 'cv_val_1.json')) as file:
   valid_json = json.load(file)
```

```python
train_coco = COCO(os.path.join(data_folder, 'cv_train_1.json'))
valid_coco = COCO(os.path.join(data_folder, 'cv_val_1.json'))
```

```python
train_num = len(train_json['images'])
valid_num = len(valid_json['images'])
total = train_num+valid_num
print('train image 개수: {} / valid image 개수: {} (ratio: {:.3}:{:.3})'.format(train_num, valid_num, train_num/total*10, valid_num/total*10))
```

```python
# train, valid 데이터셋의 각 클래스별 분포를 확인함.
categories_name = ['background', 'General trash', 'Paper', 'Paper pack', 'Metal', 'Glass', 'Plastic', 'Styrofoam', 'Plastic bag', 'Battery', 'Clothing']
train_categories = {}
for name in categories_name:
    train_categories[name] = 0

for ann in train_json['annotations']:
    category_id = ann['category_id']
    train_categories[categories_name[category_id]] += 1

valid_categories = {}
for name in categories_name:
    valid_categories[name] = 0

for ann in valid_json['annotations']:
    category_id = ann['category_id']
    valid_categories[categories_name[category_id]] += 1

fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(30, 10))
train = axs[0].bar(train_categories.keys(), train_categories.values())
for rect in train:
    height = rect.get_height()
    axs[0].text(rect.get_x()+rect.get_width()/2.0, height, str(height), ha='center', va='bottom', size=10)
valid = axs[0].bar(valid_categories.keys(), valid_categories.values())
for rect in valid:
    height = rect.get_height()
    axs[0].text(rect.get_x()+rect.get_width()/2.0, height, str(height), ha='center', va='bottom', size=10)
axs[0].set_title("The number of Each Category of Train, Valid DataSets")

ratios = []

for i, j in zip(train_categories.values(), valid_categories.values()):
    try: ratios.append(i/j)
    except ZeroDivisionError: ratios.append(0)
axs[1].barh(categories_name, ratios)
axs[0].set_title("Each Category's ratio of Train, Valid DataSets")

plt.show()
```

```python
# 하나의 이미지에 몇 개의 객체가 들어있는지 히스토그램을 그려 확인함.
train_img_objects = []

image_ids = train_coco.getImgIds()
for image_id in image_ids:
    image_info = train_coco.loadImgs(image_id)[0]
    ann_ids = train_coco.getAnnIds(imgIds=image_info['id'])
    anns = train_coco.loadAnns(ann_ids)

    train_img_objects.append(len(anns))

valid_img_objects = []

image_ids = valid_coco.getImgIds()
for image_id in image_ids:
    image_info = valid_coco.loadImgs(image_id)[0]
    ann_ids = valid_coco.getAnnIds(imgIds=image_info['id'])
    anns = valid_coco.loadAnns(ann_ids)

    valid_img_objects.append(len(anns))

print('train: min-{} max-{}'.format(min(train_img_objects), max(train_img_objects)))
print('train: min-{} max-{}'.format(min(valid_img_objects), max(valid_img_objects)))

plt.figure(figsize=(7, 5))
plt.hist(train_img_objects)
plt.hist(valid_img_objects)
plt.show()
```

```python
# 위 히스토그램 그래프에서 segmentation이 하나도 없는 이미지가 존재하는 것을 확인함.
# 따라서 해당 이미지를 체크함.
image_ids = train_coco.getImgIds()
for image_id in image_ids:
    image_info = train_coco.loadImgs(image_id)[0]
    ann_ids = train_coco.getAnnIds(imgIds=image_info['id'])
    anns = train_coco.loadAnns(ann_ids)

    file_name = image_info['file_name']
    if len(anns)==0: 
        print(file_name)
        img = np.array(Image.open(os.path.join('../data', file_name)))
        plt.figure(figsize=(8, 8))
        plt.title(file_name)
        plt.axis('off')
        plt.imshow(img)
        plt.show()
```

```python
keys = ['info', 'licenses', 'categories', 'images', 'annotations']
new_json = {}

for key in keys:
    if key != 'images':
        new_json[key]=train_json[key]
    else:
        # key == 'images'
        new_json[key] = []
        for image in train_json[key]:
            if image['file_name']=='batch_03/0702.jpg':
                continue
            new_json[key].append(image)

with open('../data/new_cv_train_1.json','w') as f:
    json.dump(new_json, f)
```
